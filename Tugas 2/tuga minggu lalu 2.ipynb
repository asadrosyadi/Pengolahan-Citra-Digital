```python
%matplotlib inline 
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
from scipy import signal
import cv2
img = cv2.imread('prambanan.jpg')
```

```python
img2 = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img2)
```

```python
bilateral = cv2.bilateralFilter(img2,9,75,75)
plt.figure()
plt.imshow(bilateral)
```

```python
gb = cv2.GaussianBlur(bilateral, (3,3), 1,1)
plt.figure()
plt.imshow(gb)
```

```python
imgx = cv2.cvtColor(gb, cv2.COLOR_BGR2GRAY)
plt.imshow(imgx, cmap = 'gray')
```

```python
ret, thresh1 = cv2.threshold(imgx, 120, 255, cv2.THRESH_BINARY)
plt.figure()
plt.imshow(thresh1, cmap = 'gray')
```
