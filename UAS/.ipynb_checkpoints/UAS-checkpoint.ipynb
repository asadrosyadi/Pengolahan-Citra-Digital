K-Means Clustering merupakan algoritma untuk menentukan cluster dalam sekumpulan data, di mana algoritma ini silakukan analisis kelompok yang mengacu pada pemartisian N objek ke dalam K kelompok (Cluster) berdasarkan nilai rata-rata (mean) terdekat

# Penerapan K-Means dalam Program

### Masukkan Gambar

```python
## read the image / membaca citra
image = cv2.imread("ub.png")
```

```python
## convert to RGB / konversi ke RGB
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
```

```python
## reshape the image to a 2D array of pixels and 3 color values (RGB)
pixel_values = image.reshape((-1, 3))
## convert to float
pixel_values = np.float32(pixel_values)
```

```python
print(image.shape)
print(pixel_values.shape)
```

```python
## define stopping criteria
criteria = (cv2.TERM_CRITERIA_EPS + cv2.TERM_CRITERIA_MAX_ITER, 100, 0.2)
```

```python
## number of clusters (K)
k = 5
_, labels, (centers) = cv2.kmeans(pixel_values, k, None, criteria, 10, cv2.KMEANS_RANDOM_CENTERS)
```

```python
## convert back to 8 bit values
centers = np.uint8(centers)

## flatten the labels array
labels = labels.flatten()
```

```python
## convert all pixels to the color of the centroids
segmented_image = centers[labels.flatten()]
```

```python
## reshape back to the original image dimension
segmented_image = segmented_image.reshape(image.shape)

#noise filter with median filter
median = cv2.medianBlur(segmented_image,5)

#open area menghilangkan noise
kernel = np.ones((5,6))
opening = cv2.morphologyEx(median, cv2.MORPH_OPEN, kernel)
```

```python
#show the image
figure_size = 15
plt.figure(figsize=(figure_size,figure_size))
plt.subplot(2,2,1),plt.imshow(image)
plt.title('Original Image')

plt.subplot(2,2,2),plt.imshow(segmented_image)
plt.title('Segmented Image when K= %i' % k)
          
plt.subplot(2,2,3),plt.imshow(median)
plt.title('Segmentation Image with Median Filtering')
plt.subplot(2,2,4),plt.imshow(opening)
plt.title('Segmented Image with Open Area')
plt.show()
```
