```python
import matplotlib, cv2
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mping

%matplotlib inline
```

```python
# menyimpan gambar dengan fungsi imread dari OpenCV
img = cv2.imread('5469.jpg')
```

```python
# membuat sebuah variabel kosong 'fig'
fig = plt.figure()
fig.add_subplot(121)
# subplot (jumlah baris, jumlah kolom, urutan)

# plotting gambar
fig.add_subplot(122)
# subplot, formatnya: (jumlah baris, jumlah kolom, urutan)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# menampilkan gambar dengan fungsi plt
plt.imshow(img)

# lihat tipe data img. disimpan sebagai apa?
print(type(img))
```

```python
#convert image to grayscale
gray_img = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)

#grayscale image represented as a 2-d array
print(gray_img)
```

```python
#have to converet grayscale back to RGB for plt.imshow()
plt.imshow(cv2.cvtColor(gray_img, cv2.COLOR_GRAY2RGB))
```
